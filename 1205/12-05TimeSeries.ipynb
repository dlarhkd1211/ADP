- 삼성전자, 구글은 정상성 X

```python
df.diff(1).dropna().plot(subplots=True, figsize=(12,5))
```

```python
import statsmodels.api as sm
```

```python
diff_df = df.diff(1).dropna()
```

```python
fit = sm.tsa.VAR(diff_df).fit(maxlags=2)
```

```python
display(fit.summary())
```

```python
fit.plot_forecast(200) 
plt.show()
```
