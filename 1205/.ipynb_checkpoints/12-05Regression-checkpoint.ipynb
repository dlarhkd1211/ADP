```python
from sklearn.preprocessing import LabelEncoder
```

```python
col = ['route_nm','now_station','next_station','now_arrive_time']
```

```python
for i in col:
    lab = LabelEncoder()
    lab.fit(df[i])
    df[i] = lab.transform(df[i])
```

```python
for i in col:
    lab = LabelEncoder()
    lab.fit(test_df[i])
    test_df[i] = lab.transform(test_df[i])
```

```python
X = df[df.columns.difference(['next_arrive_time'])]
y = df['next_arrive_time']
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV, KFold
```

```python
rf_param_grid = {
    'n_estimators': [100, 150, 200, 250],
    'max_depth': [None, 6, 9, 12],
    'min_samples_split': [0.01, 0.05, 0.1],
    'max_features': ['auto', 'sqrt'],
}
kf = KFold(random_state = 0,
          n_splits = 10,
          shuffle = True)
```

```python
rf = RandomForestRegressor()
```

```python
rf_grid_search = GridSearchCV(estimator = rf,
                          param_grid = rf_param_grid,
                          cv = kf,
                          n_jobs = -1,
                          verbose = 2)
```

```python
rf_grid_search.fit(X, y)
```

```python
pred = rf_grid_search.best_estimator_.predict(test_df)
```

```python
submission = pd.read_csv('submission_제출양식.csv')
submission['next_arrive_time'] = pred
submission.to_csv("submission_bus.csv", index=False)
```
